**import Required Libraries**

```python
import math
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

**create data**

```python
f = {'Experience':[2,4,1,5,3,9,7],
     'Salary':[42,55,40,57,50,70,90]
     } # using dict we can create the data -> but if we give dict to pandas it will create excel sheet data


df = pd.DataFrame(f)

df
```

```python
X = df['Experience'] # independent column
y = df['Salary'] # dependent column
```

- `Best Machine Learning Library -> https://scikit-learn.org/stable/`

```python
import sklearn # Machine learning Library
```

**Break the data into Training_Part and Testing_Part**

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2)
```

```python
training_part = pd.DataFrame({'X_train':X_train,'y_train':y_train})
training_part
```

```python
testing_part = pd.DataFrame({'X_test':X_test,'y_test':y_test})
testing_part
```
